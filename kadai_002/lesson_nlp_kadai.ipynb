```python
# ライブラリをインポート
from bs4 import BeautifulSoup
from urllib import request

# URL内の文章を取得する
url = 'https://www.aozora.gr.jp/cards/000148/files/2371_13943.html'
response = request.urlopen(url) # urllib.requestモジュールのurlopen関数でURLの内容を取得
soup = BeautifulSoup(response) # 取得した内容BeautifulSoupに入力
response.close()

print(soup)
```

```python
# タグの削除
# main_textのみを取得
main_text = soup.find('div', class_='main_text')

# ルビの削除
tags_to_delete = main_text.find_all(['rp', 'rt'])
for tag in tags_to_delete:
    tag.decompose()

# その他のタグの削除
main_text = main_text.get_text()

#スペース(\u3000)、改行(\n, \r)の削除
import re
main_text = re.sub(r"[\u3000 \n \r]", "", main_text)
print(main_text)

```

```python
# ライブラリをインポート
from bs4 import BeautifulSoup
from urllib import request

# ストップワードの除去（SlothLibを活用した辞書ベース）
# SlothLibからストップワードを取得する
url = 'http://svn.sourceforge.jp/svnroot/slothlib/CSharp/Version1/SlothLib/NLP/Filter/StopWord/word/Japanese.txt'
response = request.urlopen(url)
soup = BeautifulSoup(response)
response.close()

# htmlタグなどを削除
stopwords_text = soup.text

# 改行を削除してリスト形式に変換する
stopwords_list = stopwords_text.split("\r\n") # 改行をベースに単語毎に分解してリスト化
stopwords_list = [word for word in stopwords_list if word] # 空白の文字列を除去
print(stopwords_list)
```

```python
# ストップワードの除去
#　split_text_listの各単語の内、stopwords_listに載っていないものだけをresult_text_listに加える
split_text_list =  ['近頃', 'は', '大分', '方々', 'の', '雑誌', 'から', '談話', 'を', 'しろ', 'しろ', 'と', '責め', 'られて', '、', '頭が', 'がらん胴', 'に', 'なっ', 'た', 'から', '、', '当分', '品切れ', 'の', '看板', 'でも', '懸け', 'たい', 'くらい', 'に', '思って', 'います', '。']
result_text_list = list()
for split_text in split_text_list: 
  if split_text not in stopwords_list:
    result_text_list.append(split_text)

print(result_text_list)
```
